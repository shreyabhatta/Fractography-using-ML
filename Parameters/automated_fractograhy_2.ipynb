```python
import skimage.io as skimage
import matplotlib.pyplot as plt

image = skimage.imread(fname="ductile-3.jpg")

fig, ax = plt.subplots()
plt.imshow(image)
```

```python
import skimage.color

gray_image = skimage.color.rgb2gray(image)
fig, ax = plt.subplots()
plt.imshow(gray_image, cmap="gray")
```

```python
import skimage.filters

blurred = skimage.filters.gaussian(
    gray_image, sigma=(3.0,3.0), truncate=3.5)
fig, ax = plt.subplots()
plt.imshow(blurred, cmap="gray")
```

```python
thresh = skimage.filters.threshold_otsu(blurred)
binary_mask = blurred > thresh

fig, ax = plt.subplots()
plt.imshow(binary_mask, cmap="gray")
```

```python
from skimage import feature

# Compute the Canny filter for two values of sigma
edges1 = feature.canny(binary_mask)
edges2 = feature.canny(binary_mask, sigma=3)

fig, ax = plt.subplots()
plt.imshow(edges1, cmap="gray")

fig, ax = plt.subplots()
plt.imshow(edges2, cmap="gray")
```

```python
import skimage.io
import skimage.color
import skimage.filters
from skimage import feature
import numpy as np
import matplotlib.pyplot as plt


# Reading the image 
image = skimage.io.imread(fname="ductile-3.jpg")

# Convert to grayscale
gray_image = skimage.color.rgb2gray(image)

# Applying Gaussian Blur
# sigma = 1.0
blurred1 = skimage.filters.gaussian(
    gray_image, sigma=(1.0, 1.0), truncate=3.5)
blurred2 = skimage.filters.gaussian(
    gray_image, sigma=(3.0, 3.0), truncate=3.5)

#Applying threshold
thresh1 = skimage.filters.threshold_otsu(blurred1)
binary_mask1 = blurred1 > thresh1

thresh2 = skimage.filters.threshold_otsu(blurred2)
binary_mask2 = blurred2 > thresh2

#Applying Canny Edge Detection
edges11 = feature.canny(binary_mask1)
edges12 = feature.canny(binary_mask1, sigma=3)

edges21 = feature.canny(binary_mask2)
edges22 = feature.canny(binary_mask2, sigma=3)

# display results
fig, ax = plt.subplots(nrows=3, ncols=3, figsize=(48,18))

ax[0][0].imshow(image, cmap='gray')
ax[0][0].set_title(r'original', fontsize=20)

ax[0][1].imshow(blurred1, cmap='gray')
ax[0][1].set_title(r'blurred-1, $\sigma=1$', fontsize=20)

ax[0][2].imshow(blurred2, cmap='gray')
ax[0][2].set_title(r'blurred-2, $\sigma=3$', fontsize=20)

ax[1][0].imshow(binary_mask1, cmap='gray')
ax[1][0].set_title(r'threshold-1, $\sigma=3$', fontsize=20)

ax[1][1].imshow(binary_mask2, cmap='gray')
ax[1][1].set_title(r'threshold-2', fontsize=20)

ax[1][2].imshow(edges11, cmap='gray')
ax[1][2].set_title(r'Canny filter-1, $\sigma=1$', fontsize=20)

ax[2][0].imshow(edges12, cmap='gray')
ax[2][0].set_title(r'Canny filter-1, $\sigma=3$', fontsize=20)

ax[2][1].imshow(edges21, cmap='gray')
ax[2][1].set_title(r'Canny filter-2, $\sigma=1$', fontsize=20)

ax[2][2].imshow(edges22, cmap='gray')
ax[2][2].set_title(r'Canny filter-2, $\sigma=3$', fontsize=20)

# for a in ax:
#     a.axis('off')

fig.tight_layout()
plt.show()

```
